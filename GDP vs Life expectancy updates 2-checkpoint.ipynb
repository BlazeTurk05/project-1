```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
%matplotlib notebook
```

```python
data = "gapminder.tsv"
```

```python
# Read CSV
df_raw = pd.read_csv('gapminder.tsv', sep='\t')
print(df_raw.shape)
df_raw.head()
df_raw.info()
```

```python
df= df_raw.set_index(['country'])
df.head()
```

```python
df['pop'] = df['pop'] / 1000000
df.head()
```

```python
df.columns
```

```python
pop = df['pop'].tolist()
pop[:10]
df.round(2)
```

```python
gdp_cap = df['gdpPercap'].tolist()
gdp_cap[:10]
df.round(2)
```

```python
life_exp = df['lifeExp'].tolist()
life_exp[:10]
df.round(2)
```

```python
year = df['year'].tolist()
year[:10]
```

```python
#Calculate which continents have the highest and lowest gdp and life expectancy
continent_gdp = df.groupby('continent')['gdpPercap'].mean() 
continent_gdp
```

```python
continents = ['Africa','Americas','Asia','Europe','Oceania']
plt.bar(continents,continent_gdp, color='b', alpha =.5, align ='center')
xlab = "continents"
ylab = "gdp per cap"
plt.title = "High gdp vs Low gdp"
plt.xlabel (xlab)
plt.ylabel (ylab)
plt.savefig("figure.png")
```

```python
continent_lifeExp = df.groupby('continent')['lifeExp'].mean()
continent_lifeExp
```

```python
continents = ['Africa','Americas','Asia','Europe','Oceania']
plt.bar(continents, continent_lifeExp, color='b', alpha =.5, align ='center')
xlab = "continents"
ylab = "Life Expectancy"
plt.title = "Life Expectancy gdp vs Low Life Expectancy"
plt.xlabel (xlab)
plt.ylabel (ylab)
plt.savefig("figure.png")
```

```python
#Plot current DF using seaborn
sns.pairplot(df);
```

```python
#The data is currently right schewed so we need to convert values to log to better visualize data.

df['log_pop'] = np.log10(df['pop'])
df['log_gdp_per_cap'] = np.log10(df['gdpPercap'])

df = df.drop(columns = ['pop', 'gdpPercap'])
df
```

```python
#Rename columns
clean_df = df.rename(columns={'log_pop':'Population','log_gdp_per_cap':'GDP Per Capita','lifeExp':'Life Expectancy','year':'Year','continent':'Continent'})
```

```python
#Sort by continents
matplotlib.rcParams['font.size'] = 20
sns.pairplot(clean_df, hue = 'Continent');
```

```python
#Improve the plot to improve visualization
sns.pairplot(clean_df, hue = 'Continent', diag_kind = 'kde', plot_kws = {'alpha': 1, 's': 60, 'edgecolor': 'k'}, height = 4);
```

```python
#Bin time periods together
clean_df['decade'] = pd.cut(clean_df['Year'], bins = range(1950, 2010, 10))
```

```python
#Plot the time periods
sns.pairplot(clean_df, hue = 'decade', diag_kind = 'kde', vars = ['Life Expectancy', 'Population', 'GDP Per Capita'],
             plot_kws = {'alpha': 1, 's': 60, 'edgecolor': 'k'}, height = 4);
```
